```python
# Error
weight = 0.1
lr = 0.01

def neural_network(input, weight):
    prediction = input * weight
    return prediction

# PREDICT: Making a prediction and evaluating the error
number_of_toes = [8.5]
win_or_lose_binary = [1] # (won!!)

input = number_of_toes[0]
true = win_or_lose_binary[0]

pred = neural_network(input, weight)

error = (pred - true) ** 2
print(error)

p_up = neural_network(input, weight + lr)

e_up = (p_up - true) ** 2

print("P up", e_up)

p_dn = neural_network(input, weight - lr)

e_dn = (p_dn - true) ** 2

print("P down", e_dn)

## Compare and learn: Comparing the errors and setting the new weight
print("Weight before compare and learn", weight)
if(error > e_dn or error > e_up):
    if (e_dn < e_up):
        weight = weight - lr
    else:
        weight = weight + lr

print("Weight after compare and learn", weight)
```

## Changes with the simples form of learning - Hot or Cold

```python
weight = 0.5
input = 0.5
goal_prediction = 0.8

step_amount = 0.8

for iteration in range(1101):
    prediction = input * weight

    error = (prediction - goal_prediction) ** 2

    print("Error:", str(error) + " Prediction:" + str(prediction))

    up_prediction = input * (weight + step_amount)
    up_error = (goal_prediction - up_prediction) ** 2

    down_prediction = input * (weight - step_amount)
    down_error = (goal_prediction - down_prediction) ** 2

    if(down_error < up_error):
        weight = weight - step_amount

    if(down_error > up_error):
        weight = weight + step_amount
```

# Calculating both direction and amount from error

```python
weight = 0.5
goal_pred = 0.8
input = 0.5

for iteration in range(20):
    pred = input * weight
    error = (pred - goal_pred) ** 2
    direction_and_amount = (pred - goal_pred) * input
    weight = weight - direction_and_amount

    print("Error:" + str(error) + " Prediction:" + str(pred))
```

# One iteration of gradient descent

```python
weight = 0.1
alpha = 0.01

def neural_network(input, weight):
    prediction = input * weight
    return prediction

number_of_toes = [8.5]
win_or_lose_binary = [1]

input = number_of_toes[0]
goal_pred = win_or_lose_binary[0]

pred = neural_network(input, weight)

# Forces the raw error to be positive by multiplying
# it by itself. Negative error won't make sense.
error = (pred - goal_pred) ** 2

# Node delta
delta = pred - goal_pred

# Weight delta
weight_delta = input * delta

alpha = 0.01

weight -= weight_delta * alpha

```

# Learning is just reducing error

```python
weight, goal_pred, input = (0.0, 0.8, 0.5)

for iteration in range(4):
    pred = input * weight
    error = (pred - goal_pred) ** 2
    delta = pred - goal_pred
    weight_delta = delta * input
    weight = weight - weight_delta
    print("Error:" + str(error) + " Prediction:" + str(pred))
```
